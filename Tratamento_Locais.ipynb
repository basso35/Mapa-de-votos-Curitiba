# Tratamento dos locais de votação

Este notebook contém o tratamento da base com a geolocalização dos locais de votação de Curitiba.

Na query realizada para extração dos dados, foram extraidas apenas os locais pertencentes ao municpio de Curitiba

```python
#importação de bibliotecas
import pandas as pd
import numpy as np
```

```python
#criando um df apenas com localizações unicas
df1 = df[['zona','secao','melhor_urbano','melhor_rural','tse_recente']]
df2 = df1.drop_duplicates()
```
